```python
import datetime
def convert(date_time):
    format1 = '%m/%d/%Y %H:%M'
    dat = pd.to_datetime(date_time, format=format1).dt
    return dat.month, dat.day, dat.hour, dat.minute
```

```python
def qc_score_flag():
    for i in tqdm(range(df.shape[0])):
        if(df['R_flag'][i] != -1):
            qc_flag = 'R'+str(df['R_flag'][i])
        if(df['I_flag'][i] != -1):
            qc_flag += 'I'+str(df['I_flag'][i])
        if(df['Z_flag'][i] != -1):
            qc_flag += 'Z'+str(df['Z_flag'][i])
        if(df['B_flag'][i] != -1):
            qc_flag += 'B'+str(df['B_flag'][i])

        qc_score = -1
        if(qc_flag in qc0):
            qc_score = 0
        elif(qc_flag in qc1):
            qc_score = 1
        elif(qc_flag in qc2):
            qc_score = 2
        elif(qc_flag in qc3):
            qc_score = 3
        df['qc_score'][i] = qc_score

qc_score_flag()
```
